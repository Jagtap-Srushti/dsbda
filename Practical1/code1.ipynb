```python
# calculate the mean value for 'stroke' column
avg_stroke = df['stroke'].astype('float').mean(axis = 0)
print("Average of stroke : ",avg_stroke)

# replace NaN by mean value in 'stroke' column
df['stroke'] = df['stroke'].replace(np.nan, avg_stroke)
```

```python

# Calculate the mean for the horsepower
avg_hp = df['horsepower'].astype('float').mean(axis = 0)
print("Average of horsepower : ",avg_hp)
```

```python
# replace NaN by mean value
df['horsepower'].replace(np.nan,avg_hp,inplace = True)
```

```python
# calculate the mean value for 'peak-rpm' column
avg_rpm = df['peak-rpm'].astype('float').mean(axis = 0)
print(r"Average of 'peak-rpm' : ",avg_rpm)
```

```python
df['peak-rpm'] = df['peak-rpm'].replace(np.nan, avg_rpm)
```

```python

df['num-of-doors'].value_counts()
```

```python

df['num-of-doors'].value_counts().idxmax()
```

```python
# replace the missing 'num-of-doors' values by the most frequent
df['num-of-doors'] = df['num-of-doors'].replace(np.nan, "four")

# simply drop whole row with NaN in "horsepower -binned" column
df.dropna(subset = ["horsepower-binned"],axis=0,inplace=True)

# reset index, because we dropped two rows
df.reset_index(drop=True,inplace = True)
```

```python
df.isnull().sum()
```

```python

df['city-L/100km'] = 235/df['city-mpg']
df.head()
```

```python

df['highway-L/100km'] = 235/df["highway-mpg"]
df.head()
```

```python
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
```

```python

df['height'] = df['height']/df['height'].max()
df[["length","width","height"]].head()
```

```python

df.columns
```

```python

df['aspiration'].value_counts()
```

```python

dummy_variable_1 = pd.get_dummies(df["aspiration"])
dummy_variable_1.head()
```

```python

df = pd.concat([df,dummy_variable_1],axis = 1)
df.drop('aspiration',axis = 1,inplace=True)
```

```python

df.head()
```

```python

# Horsepower
df['horsepower']=df['horsepower'].astype(float,copy = True)
```

```python
import matplotlib.pyplot as plt
plt.hist(df['horsepower'])
plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower-bins")
```

```python

bins = np.linspace(min(df['horsepower']),max(df['horsepower']),4)
bins
```

```python

group_names = ['Low', 'Medium', 'High']
```

```python

df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head(20)
```

```python

df["horsepower-binned"].value_counts()
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```
