```python
data['Roll no '] = data['Roll no '].str[-3:]

data.head()
```

```python
# -------------------------------------------------
# 6. Check and Handle Inconsistencies
# -------------------------------------------------
# Example:
# CGPA values cannot be 95 or 0
# Convert 95 -> 9.5

data['First year:   Sem 1'] = data['First year:   Sem 1'].replace(95, 9.5)
data['First year:   Sem 2'] = data['First year:   Sem 2'].replace(95, 9.5)

print("\nInconsistencies handled successfully.")
```

```python

sns.boxplot(y=data['First year:   Sem 1'])
```

```python
# -------------------------------------------------
# 8. Remove Outliers Using IQR Method
# -------------------------------------------------

Q1 = data['First year:   Sem 1'].quantile(0.25)
Q3 = data['First year:   Sem 1'].quantile(0.75)

IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

data = data[
    (data['First year:   Sem 1'] >= lower_limit) &
    (data['First year:   Sem 1'] <= upper_limit)
]

print("\nOutliers Removed Successfully")
```

```python
sns.boxplot(y=data['First year:   Sem 1'])
```

```python

sns.boxplot(y=data['First year:   Sem 2'])
```

```python
# -------------------------------------------------
# 8. Remove Outliers Using IQR Method
# -------------------------------------------------

Q1 = data['First year:   Sem 2'].quantile(0.25)
Q3 = data['First year:   Sem 2'].quantile(0.75)

IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

data = data[
    (data['First year:   Sem 2'] >= lower_limit) &
    (data['First year:   Sem 2'] <= upper_limit)
]

print("\nOutliers Removed Successfully")
```
